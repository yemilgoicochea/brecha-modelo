```python
import pandas as pd
import re

# 1. Cargar CSV
# Cambia 'input.csv' por el nombre de tu archivo
df = pd.read_csv("proyecto_limpio.csv", delimiter=",")
```

```python

# Lista de palabras a eliminar
palabras = ["urbano", "urbana", "rural"]

# Expresión regular para encontrarlas (palabra exacta)
pattern = r"\b(" + "|".join(palabras) + r")\b"

# 2. Crear columna "BRECHA_limpio2" eliminando esas palabras
df["BRECHA_limpio2"] = df["BRECHA_limpio"].str.replace(pattern, "", regex=True)

# Limpiar espacios múltiples generados por la eliminación
df["BRECHA_limpio2"] = df["BRECHA_limpio2"].str.replace(r"\s+", " ", regex=True).str.strip()

# 3. Extraer las palabras eliminadas para la columna "URBANO_RURAL"
df["URBANO_RURAL"] = df["BRECHA_limpio"].apply(
    lambda x: ", ".join(re.findall(pattern, x.lower()))
)

# 4. Exportar archivo final
df.to_csv("proyecto_limpio_2.csv", index=False, sep=",")
```
