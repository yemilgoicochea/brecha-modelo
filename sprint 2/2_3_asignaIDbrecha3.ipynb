```python
import pandas as pd

# 1. Leer el CSV
# Ajusta la ruta si es necesario. Si tu archivo usa coma como separador, cambia sep=";" por sep=",".
input_file = "proyecto_limpio_3.csv"
output_file = "proyecto_limpio_4.csv"

df = pd.read_csv(input_file, sep=",")

# 2. Asegurar que la columna ID exista; si no existe, crearla con 0
if "ID" not in df.columns:
    df["ID"] = 0

# 3. Cambiar las filas con columna ID == 520 a 1
df.loc[df["ID"] == 520, "ID"] = 1

# 4. Definir el mapeo de palabras clave en BRECHA_limpio -> ID
mapeo_brechas = {
    "agua potable": 2,
    "aguas residuales": 3,
    "m2 de espacios": 4,
    "movilidad": 5,
    "educacion inicial": 6,
    "educacion primaria": 7,
    "educacion secundaria": 8
}

# 5. Aplicar el mapeo solo a filas donde ID == 0
for texto_buscar, nuevo_id in mapeo_brechas.items():
    # Filas donde ID es 0 y la columna BRECHA_limpio contiene el texto
    mask = (df["ID"] == 0) & df["BRECHA_limpio"].astype(str).str.contains(texto_buscar, case=False, na=False)
    df.loc[mask, "ID"] = nuevo_id

# 6. Guardar el CSV resultante
df.to_csv(output_file, sep=";", index=False)

print("Archivo generado:", output_file)
```
