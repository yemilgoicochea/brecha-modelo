```python
import spacy
from spacy.matcher import Matcher
import re

# ------------------------------------------------------------
# Load Spanish NLP model
# ------------------------------------------------------------
nlp = spacy.load("es_core_news_lg")
```

```python
# Define rule-based matchers for each category
# ------------------------------------------------------------
matcher_alcantarillado = Matcher(nlp.vocab) #tt: add lower atrib
matcher_agua = Matcher(nlp.vocab)
matcher_educacion = Matcher(nlp.vocab)
matcher_pistas = Matcher(nlp.vocab)
matcher_esparcimiento = Matcher(nlp.vocab)
```

```python
# ------------------------------------------------------------
# Load patterns manually
# ------------------------------------------------------------

alcantarillado_patterns = [
    [{"LOWER": "saneamiento"}],
    [{"LOWER": "alcantarillado"}],
    [{"LOWER": "desagüe"}],
    [{"LOWER": "desague"}],
    [{"LOWER": "ptar"}],
    [{"LOWER": "drenaje"}],
    [{"LOWER": "aguas residuales"}],
    [{"LOWER": "aguas servidas"}],
    [{"LOWER": "ubs"}],
    [{"LOWER": "unidades basicas de saneamiento"}],
    [{"LOWER": "disposicion sanitaria"}],
    [{"LOWER": "excretas"}],
]

agua_patterns = [
    [{"LOWER": "bombeo"}],
    [{"LOWER": "pozo"}],
    [{"LOWER": "reservorio"}],
    [{"LOWER": "linea de conduccion"}],
    [{"LOWER": "sistema de conduccion"}],

    [{"LOWER": "agua potable"}],

    ]

educacion_patterns = [
    [{"LOWER": "infraestructura educativa"}],
    [{"LOWER": "nivel inicial"}],
    [{"LOWER": "nivel primario"}],
    [{"LOWER": "nivel secundario"}],
    [{"LOWER": "educacion basica regular"}],
    [{"LOWER": "institucion educativa"}],
    [{"LOWER": "servicio educativo"}],
    [{"LOWER": "infraestructura educativa complementaria"}],
    [{"LOWER": "aula"}],
    [{"LOWER": "comedor escolar"}],
    [{"LOWER": "educacion digital"}],
    [{"LOWER": "implementacion de tic"}],
]

pistas_patterns = [
    [{"LOWER": "pista"}],
    [{"LOWER": "vereda"}],
    [{"LOWER": "movilidad urbana"}],
    [{"LOWER": "infraestructura vial"}],
    [{"LOWER": "transitabilidad"}],
    [{"LOWER": "vehicular"}],
    [{"LOWER": "peatonal"}],
    [{"LOWER": "vias"}],
    [{"LOWER": "vias urbanas"}],
    [{"LOWER": "vias principales"}],
    [{"LOWER": "vial"}],

    ]

esparcimiento_patterns = [
    [{"LOWER": "espacio publico recreativo"}],
    [{"LOWER": "espacios publicos recreativos"}],
    [{"LOWER": "infraestructura deportiva"}],
    [{"LOWER": "infraestructura cultural"}],
    [{"LOWER": "recreacion"}],
    [{"LOWER": "parque"}],
    [{"LOWER": "plaza"}],
    [{"LOWER": "deportivo"}],
    [{"LOWER": "comunal"}],
    [{"LOWER": "local comunal"}],
    [{"LOWER": "cultural"}],
    [{"LOWER": "recreativos"}],
    [{"LOWER": "servicio de recreacion"}],
    [{"LOWER": "areas verdes"}],
    [{"LOWER": "espacio publico urbano"}],
]

matcher_alcantarillado.add("ALCANTARILLADO", alcantarillado_patterns)
matcher_agua.add("AGUA", agua_patterns)
matcher_educacion.add("EDUCACION", educacion_patterns)
matcher_pistas.add("PISTAS", pistas_patterns)
matcher_esparcimiento.add("ESPARCIMIENTO", esparcimiento_patterns)
```

```python
# ------------------------------------------------------------
# Function to classify a single project title
# ------------------------------------------------------------
def classify_project(title):
    doc = nlp(title)

    if matcher_alcantarillado(doc):
        return "Servicio de Alcantarillado"
    if matcher_agua(doc):
        return "Servicio de Agua potable"
    elif matcher_educacion(doc):
        return "Servicio de Educación"
    elif matcher_pistas(doc):
        return "Servicio de Transporte (Pistas y Veredas)"
    elif matcher_esparcimiento(doc):
        return "Servicio de Esparcimiento"
    else:
        return "El proyecto no cierra brechas"
```

```python
# ------------------------------------------------------------
print("--- Project Classifier  ---")


while True:
     user_title = input("\nIngrese el título del proyecto (o 'salir' para terminar): ")
     if user_title.lower() == 'salir':
         break
     if not user_title.strip():
         print("Por favor ingrese un título de proyecto.")
         continue

     user_title_normalized = re.sub(r'[^\w\sáéíóúüñ]', ' ', user_title)
     category = classify_project(user_title_normalized)
     print(f"Categoría detectada: {category}")
```
